# ðŸ“š Book Sales EDA

This project performs exploratory data analysis (EDA) on a cleaned book sales dataset. We analyze ratings, genres, pricing, publishing years, and more to uncover insights.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## ðŸ“¥ Data Loading & Preview

```python
df = pd.read_csv('Books_Data_Clean.csv')
```

```python
df.head()
```

```python
df.describe()
```

## ðŸ§¹ Data Cleaning & Preparation

```python
df=df[df["Publishing Year"]>1900]
```

```python
df.isna().sum() 
```

```python
df.dropna(subset="Book Name", inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.nunique()
```

```python
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of Publishing Year")
plt.show()
```

```python
df["genre"].value_counts().plot(kind='bar')
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Distribution of Books by Genre")
plt.show()
```

## ðŸ“Š Exploratory Data Analysis

```python
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending=False)
```

```python
sns.boxplot(x="genre", y="Book_average_rating", data=df)
plt.xlabel("Genre")
plt.ylabel("Book Rating Count")
plt.title("Book Rating Count by Genre")
plt.show()
```

```python
plt.scatter(df["sale price"], df["units sold"])
plt.xlabel("Sale Price")
plt.ylabel("Units Sold")
plt.title("Sale Price vs Units Sold")
plt.show()
```

```python
language_counts = df['language_code'].value_counts()
```
